```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv('Shark Tank India Dataset.csv')
```

```python
#total no. of episode
len(data['episode_number'].unique())
```

```python
#total no. of pitch 
len(data['pitch_number'].unique())
```

```python
#total deal done
data['deal'].sum()
a=(data['deal']==1).sum()
r=(data['deal']==0).sum()
print ("Accept:",a)
print ("Reject:",r)
```

```python
#no. of deals in each episode
episode_deal_values = data.groupby('episode_number')['deal'].sum()
for episode_number, deal in episode_deal_values.items():
    print(f"Episode {episode_number} : No of deals {deal}")
plt.figure(figsize=(10, 6))
episode_deal_values.plot(kind='line', color='green',marker='o')
plt.title('Number of deals in each episode')
plt.xlabel('Episode')
plt.ylabel('Number of Deals')
plt.xticks(rotation=360)
plt.show()
```

```python
#Highest Pitch Ask Amount 
max_pitch_ask = data.loc[data['pitcher_ask_amount']>100,['brand_name','pitcher_ask_amount']]
print(max_pitch_ask)
plt.figure(figsize=(10,6 ))
max_pitch_ask.plot(kind='bar', color='black')
plt.title('Highest Pitch Ask Amount')
plt.xlabel('Pitch number')
plt.ylabel('Pitcher Ask Amount')
plt.xticks(rotation=360)
plt.show()
```

```python
#Least Pitch Ask Amount 
min_pitch_ask = data.loc[data['pitcher_ask_amount']<20,['brand_name','pitcher_ask_amount']]
print(min_pitch_ask)
plt.figure(figsize=(18, 7))
min_pitch_ask.plot(kind='bar', color='tan')
plt.title(' Least Pitch Ask Amount')
plt.xlabel('Pitch number')
plt.ylabel('Pitcher Ask Amount')
plt.xticks(rotation=360)
plt.show()
```

```python
#ask valuation vs deal valuation
plt.figure(figsize= (200,100))
data.plot(x = 'brand_name', y = ['ask_valuation','deal_valuation'], kind="bar", figsize=(25,26))
plt.xlabel('Sharks', fontsize = 40)
plt.ylabel('Amount in Lakhs', fontsize = 40)
plt.title('Ask Valuation vs Deal Valuation', fontsize = 40)
```

```python
#Total investment by each shark
shark_columns = ['ashneer_deal', 'anupam_deal', 'aman_deal', 'namita_deal', 'vineeta_deal', 'peyush_deal', 'ghazal_deal']
total_shark_invested = data[shark_columns].multiply(data['amount_per_shark'], axis=0).sum()
print(total_shark_invested )
plt.figure(figsize=(10, 6))
total_shark_invested .sort_values().plot(kind='bar', color='olive')
plt.title(' Total Investment by each Shark')
plt.xlabel('Shark')
plt.ylabel('Total Investment')
plt.xticks(rotation=360)
plt.show()
```

```python
#total equity received to each shark
shark_columns = ['ashneer_deal', 'anupam_deal', 'aman_deal', 'namita_deal', 'vineeta_deal', 'peyush_deal', 'ghazal_deal']
total_shark_equity = data[shark_columns].multiply(data['equity_per_shark'],axis=0).sum()
print(total_shark_equity)
plt.figure(figsize=(10, 6))
total_shark_equity.sort_values().plot(kind='bar', color='lightcoral')
plt.title(' Total Equity by each Shark')
plt.xlabel('Shark')
plt.ylabel('Total Equity')
plt.xticks(rotation=360)
plt.show()
```

```python
#No. of deals by each shark
#investrment of sharks in no. of brands
shark=['ashneer_deal','anupam_deal','aman_deal','namita_deal','vineeta_deal','peyush_deal','ghazal_deal']
total_deal=data[shark].sum()
print(total_deal)
plt.figure(figsize=(10, 6))
total_deal.sort_values().plot(kind='barh', color='cyan')
plt.title(' Total number of deals by each Shark')
plt.xlabel('Shark')
plt.ylabel('Number of Deals')
plt.xticks(rotation=360)
plt.show()
```

```python
#Total attendance of each Shark
shark=['ashneer_present','anupam_present','aman_present','namita_present','vineeta_present','peyush_present','ghazal_present']
total_attendance=data[shark].sum()
print (total_attendance)
plt.figure(figsize=(12, 6))
total_attendance.sort_values().plot(kind='bar', color='grey')
plt.title(' Total attendance of each Shark')
plt.xlabel('Shark')
plt.ylabel('Total Attendance')
plt.xticks(rotation=360)
plt.show()
```

```python
#Success rate of Pitches
Success=data['deal'].value_counts()
plt.figure(figsize=(5,5))
plt.pie(Success,labels=['Successful','Not Successful'],autopct='%1.1f%%',colors=['lightcoral','skyblue'])
plt.title('Success rate of pitches')
plt.show()
```

```python
#Total investment per episode
total_investments=data.groupby('episode_number')['deal_amount'].sum()
print(total_investments)
plt.figure(figsize=(10, 6))
total_investments.plot(kind='line', color='green',marker='o')
plt.title('Investment over time')
plt.xlabel('Episode')
plt.ylabel(' Deal Amount')
plt.xticks(rotation=360)
plt.show()
```

```python
#brand got same valuation as they requested
matched_valuation_data = data[data['ask_valuation'] == data['deal_valuation']]
print("Brands with matched valuations:")
print(matched_valuation_data[['brand_name', 'ask_valuation', 'deal_valuation']])
```
